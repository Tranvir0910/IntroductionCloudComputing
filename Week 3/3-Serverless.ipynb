### **Serverless Computing**

#### **Serverless**

- Serverless là một phương pháp tính toán giúp giảm bớt trách nhiệm về quản lý cơ sở hạ tầng thông thường như tự động mở rộng, lập lịch, vá lỗi và cung cấp các ngăn xếp ứng dụng cho các nhà cung cấp đám mây, giúp nhà phát triển tập trung thời gian và nỗ lực vào mã nguồn và logic kinh doanh cụ thể cho ứng dụng hoặc quy trình của họ.

- Serverless **không có nghĩa là không có máy chủ**, chỉ là **quản lý máy chủ vật lý hoặc ảo được loại bỏ khỏi người dùng.** 

- Môi trường tính toán serverless phân bổ tài nguyên khi cần thiết cho các ứng dụng. 

1. **Mô hình không có máy chủ:**
   - Không đòi hỏi việc cấp phát máy chủ.
   - Không yêu cầu cài đặt ngăn xếp ứng dụng và phần mềm.
   - Không cần vận hành cơ sở hạ tầng bởi nhà phát triển.

2. **Chạy mã khi được yêu cầu:**
   - Mã chỉ chạy khi có yêu cầu.
   - Tự động mở rộng theo số lượng yêu cầu được phục vụ.

3. **Thanh toán theo sử dụng:**
   - Người dùng chỉ trả tiền cho tài nguyên đang sử dụng.
   - Không trả tiền cho khả năng chờ đợi.

4. **Hiệu quả và Trừu tượng hóa cơ sở hạ tầng:**
   - Mã được thực thi dưới dạng các hàm cá nhân.
   - Mỗi hàm chạy trong một container không lưu trạng thái.
   - Trừu tượng hóa cơ sở hạ tầng khỏi nhà phát triển.

5. **Không cần ngữ cảnh trước đó:**
   - Không yêu cầu ngữ cảnh thực thi trước đó để phục vụ yêu cầu.

6. **Mô hình linh hoạt và Tạo phiên bản mới:**
   - Linh hoạt với việc tạo và triển khai hàm cá nhân.
   - Mỗi yêu cầu mới đều gọi một phiên bản mới của hàm.

#### **A Scenario**

- Hãy xem xét một kịch bản, ví dụ, bạn có thể có một nền tảng không có máy chủ giữa phía trước của trang web và tầng lưu trữ của bạn chạy các hàm cá nhân, ứng dụng không có máy chủ có thể dịch các tệp văn bản và lưu trữ chúng trong dịch vụ lưu trữ dựa trên đám mây. 

- Sử dụng phía trước của trang web, bạn gửi các tệp văn bản đến ứng dụng không có máy chủ. Ứng dụng tạo ra bản dịch bằng nhiều ngôn ngữ khác nhau và sau đó lưu trữ những tệp được dịch này trong Dịch vụ Lưu trữ Đám mây và gửi lại các liên kết đến bạn.

#### **Serverless computing services**

- Các dịch vụ tính toán serverless quan trọng hiện nay bao gồm **IBM Cloud Functions (dựa trên Apache OpenWhisk), AWS Lambda và Microsoft Azure Functions**.

- Lưu ý rằng serverless có thể không phải là lựa chọn phù hợp cho tất cả ứng dụng hoặc tình huống. 

- Bạn cần đánh giá đặc tính của ứng dụng và đảm bảo rằng ứng dụng phù hợp với các mô hình kiến trúc serverless.

#### **Serverless may not be the best fit**

- Ứng dụng thích hợp với kiến trúc serverless bao gồm các đặc tính sau:

    - Các **hàm ngắn, không lưu trạng thái chạy trong vài giây hoặc phút**.

    - Các **công việc theo mùa với đỉnh và đáy biến động**.

    - **Dữ liệu sản xuất với thời gian nghỉ quá lớn**.

    - Xử lý dựa trên sự kiện hoặc xử lý yêu cầu không đồng bộ để triển khai các trường hợp sử dụng.

    - Microservices có thể được xây dựng dưới dạng các hàm không lưu trạng thái.

#### **Use Cases**

- Kiến trúc serverless rất phù hợp cho các trường hợp **sử dụng xung quanh xử lý dữ liệu và sự kiện, IoT, microservices và backend cho ứng dụng di động**. 

- Do **khả năng tự động mở rộng, triển khai nhanh chóng** và **mô hình giá cả không thay đổi khi không hoạt động**, **hỗ trợ kiến trúc microservices đã trở thành một trong những trường hợp sử dụng phổ biến nhất của tính toán serverless ngày nay**.

- Serverless rất phù hợp để làm việc với: 

    - Dữ liệu có cấu trúc
    - Âm thanh
    - Hình ảnh và video trong các nhiệm vụ như làm giàu dữ liệu, biến đổi, xác nhận và làm sạch, xử lý PDF, làm chuẩn âm thanh, tạo hình ảnh thu nhỏ và chuyển đổi video. 

- Các nhiệm vụ song song như **tìm kiếm và xử lý dữ liệu, xử lý gen cũng rất phù hợp để chạy trên môi trường serverless**. 

- Serverless cũng rất phù hợp để làm việc với tất cả **các loại dữ liệu dòng, bao gồm dữ liệu doanh nghiệp, dữ liệu cảm biến IoT, dữ liệu nhật ký và dữ liệu thị trường tài chính**.

#### **Challenges**

1. **Các khối công việc serverless:** Thiết kế để mở rộng lên và giảm xuống phản hồi với công việc.

2. **Đối với các công việc có quá trình chạy lâu:** Quản lý môi trường máy chủ truyền thống có thể đơn giản và hiệu quả chi phí hơn.

3. **Phụ thuộc vào nhà cung cấp:** Kiến trúc ứng dụng serverless có thể bị ràng buộc với nhà cung cấp, đặc biệt là liên quan đến các khả năng của nền tảng như xác thực, tự động mở rộng, giám sát hoặc quản lý cấu hình.

4. **Khởi đầu từ không:** Vì kiến trúc serverless mở rộng lên và giảm xuống phản hồi với công việc, đôi khi chúng cần bắt đầu từ không để phục vụ yêu cầu mới.

5. **Độ trễ trong ứng dụng tài chính:** Đối với một số ứng dụng, độ trễ này không ảnh hưởng nhiều; tuy nhiên, đối với ứng dụng tài chính có độ trễ thấp, độ trễ này có thể không chấp nhận được.

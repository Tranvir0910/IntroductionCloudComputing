### **Hybrid Multi-Cloud**

#### **Hybrid cloud**

- Môi trường tính toán kết hợp (hybrid cloud), là **một môi trường tính toán kết nối đám mây riêng tư on-premise của tổ chức và đám mây công cộng của bên thứ ba thành một hạ tầng đơn để chạy ứng dụng của tổ chức**.

#### **Multi-Cloud**

- Multi-cloud là **một chiến lược sử dụng đám mây kết hợp nhiều mô hình đám mây từ các nhà cung cấp dịch vụ khác nhau**, bao gồm **đám mây công cộng, riêng tư và quản lý, trên cả hạ tầng, nền tảng hoặc dịch vụ phần mềm**.

#### **1. Scaling Đám Mây (Cloud Scaling)**

- Bắt đầu tại điểm cơ bản, đa số chúng ta có thể quen với việc mở rộng đám mây - một trong những lý do chính để chuyển đến đám mây. 

- Ví dụ, một dịch vụ giao hoa có khả năng phục vụ một lượng người dùng nhất định. Họ có cơ sở hạ tầng on-premise có thể đáp ứng tải người dùng cụ thể. Tuy nhiên, để đáp ứng những đỉnh cao đó, họ có thể mở rộng kiến trúc on-premise, nhưng điều này đến với chi phí và chi phí duy trì. 

- Thay vào đó, họ có thể tận dụng đám mây để mở rộng phản ứng vào tải và tự động giảm quyền sử dụng tài nguyên khi chúng không còn cần thiết.



#### **2. Composite Cloud**

- Mô hình này liên quan đến ứng dụng phân tán trên nhiều môi trường đám mây. 

- Ví dụ, dịch vụ giao hoa có thể chạy ba thành phần chính trên cơ sở hạ tầng on-premise: Giao diện người dùng Web (Web UI), các API thanh toán (billing APIs) và một khung công cụ thưởng (rewards framework). 

- Tuy nhiên, họ có thể nhận thấy rằng hệ thống gặp vấn đề tại Bắc Mỹ vào những dịp như Ngày Cựu chiến binh hoặc Lễ tạ ơn. Vì vậy, họ quyết định sử dụng kiến trúc đám mây kết hợp hoặc đám mây đa nhà cung cấp bằng cách xây dựng ứng dụng trên nhiều môi trường đám mây. 

- Họ di chuyển thanh toán và khả năng giao diện người dùng (UI) đến một nền tảng đám mây ở Bắc Mỹ, giúp họ mở rộng theo yêu cầu của thị trường Mỹ, trong khi các phần ở châu Âu vẫn được duy trì. Điều này cho phép họ mở rộng từng phần một cách linh hoạt, tận dụng đối với các ngày lễ cụ thể.



#### **3. Hiện Đại Hóa (Modernization) - Ngành Hàng Hàng Không**

- Trong ngành hàng không, việc hiện đại hóa có thể làm thay đổi trải nghiệm người dùng. Mặc dù nhiều ứng dụng doanh nghiệp (khoảng 80%) vẫn đặt tại nơi, một số công ty hàng không đã tạo ứng dụng di động để cải thiện trải nghiệm của hành khách. 

- Ứng dụng di động này có thể chạy trên nền tảng đám mây công cộng và tương tác với dịch vụ đặt chỗ, mang lại trải nghiệm mới cho người dùng. Một bước tiến xa hơn là sử dụng đám mây để tạo tính năng đề xuất cho hành khách khi chuyến bay bị chậm trễ. 

- Điều này kết nối với dịch vụ di động, cho phép hành khách đặt chuyến bay mới ngay khi thông báo trễ được đề xuất hoặc xảy ra. Điều này không chỉ cải thiện hiệu suất tài chính cho ngành hàng không mà còn làm hài lòng hành khách.

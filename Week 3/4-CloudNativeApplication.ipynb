### **Cloud Native Application**

- Một **Cloud Native Application** là một ứng dụng được phát triển từ đầu để chỉ hoạt động trong môi trường đám mây hoặc là một ứng dụng hiện tại đã được tái cấu trúc và cấu hình lại với các nguyên tắc native cloud. 

- Một ứng dụng native cloud bao gồm các microservices hoạt động cùng nhau như một thể để tạo thành một ứng dụng, nhưng mỗi microservice có thể được **mở rộng và cập nhật độc lập thông qua quy trình tự động hóa và điều phối.** 

- Các microservices này thường được **đóng gói trong các container**, đơn vị thực thi phần mềm trong đó mã ứng dụng được đóng gói cùng với thư viện và phụ thuộc của nó để có thể chạy từ bất kỳ đâu. 

- Sự độc lập cho phép cải thiện đa dạng, lặp đi lặp lại của các ứng dụng native cloud mà không làm gián đoạn trải nghiệm của người dùng cuối. 

- Ứng dụng native cloud khác với các ứng dụng truyền thống, hoặc monolithic, được xây dựng từ một phần mềm lớn; các ứng dụng kết hợp chặt chẽ giao diện người dùng, tầng logic doanh nghiệp và tầng dữ liệu.

- Hãy xem ví dụ về cách một ứng dụng native cloud có thể được sử dụng trên trang web du lịch. Mỗi chủ đề được trang web bao gồm - chuyến bay, khách sạn, ô tô, ưu đãi - là một microservice riêng biệt. 

- Mỗi microservice có thể triển khai các tính năng mới độc lập với các microservice khác. Các ưu đãi và giảm giá cũng có thể mở rộng độc lập. 

- Trong khi trang web du lịch được trình bày cho khách hàng như một thể thống nhất, mỗi microservice vẫn độc lập và có thể được mở rộng hoặc cập nhật khi cần mà không ảnh hưởng đến các dịch vụ khác.

- Dù tạo một ứng dụng native cloud mới hoặc hiện đại hóa một ứng dụng hiện tại, các nhà phát triển tuân theo một bộ nguyên tắc phát triển nhất quán: 

    - Theo đuổi phương pháp kiến trúc microservices bằng cách phân rã ứng dụng thành các microservices chức năng duy nhất. 
    
    - Dựa vào container để đạt được sự linh hoạt, mở rộng và di động tối đa.

Thụ động phương pháp Agile để tăng tốc quá trình tạo ra và cải thiện thông qua các bản cập nhật nhanh chóng, lặp lại dựa trên phản hồi của người dùng.

**Cloud Infrastructure (Hạ tầng Đám mây)**

- Bao gồm hạ tầng cá nhân, công cộng và doanh nghiệp. 

- Các ứng dụng cloud native có thể áp dụng cho các tình huống đa đám mây và đa đám mây.

**Scheduling and Orchestration Layer (Lớp Lên lịch và Orchestration)**

- Bao gồm các lớp điều khiển như Kubernetes, quản lý lên lịch và orchestration.

**Application and Data Services Layer (Lớp Dịch vụ Ứng dụng và Dữ liệu)**

- Bao gồm dịch vụ hỗ trợ ứng dụng và tích hợp mã nguồn ứng dụng với các dịch vụ hiện có có sẵn trên các đám mây khác nhau hoặc thậm chí on-premise.

**Application Runtimes (Thời gian chạy Ứng dụng)**

- Truyền thống được biết đến là middleware. 

- Đây là nơi mà ứng dụng code được thiết kế, xây dựng và triển khai khác biệt so với ứng dụng monolithic.

**Benefits of Cloud Native Apps (Lợi ích của Ứng dụng Cloud Native)**

- Cung cấp sự đổi mới, linh hoạt kinh doanh và quan trọng nhất là từ góc độ công nghệ, sự hóa chất của bộ giải pháp.

**Use Cases (Các Trường Hợp Sử Dụng)**

- Ứng dụng nguyên bản của đám mây thích hợp cho mọi thứ, đặc biệt là các ứng dụng có thiết kế và phương pháp nguyên bản của đám mây.

**Challenges (Thách thức)**

- Các công việc của serverless được thiết kế để mở rộng lên và giảm xuống phản hồi với công việc. 

- Tuy nhiên, đối với các công việc có quá trình chạy lâu, quản lý môi trường máy chủ truyền thống có thể đơn giản và hiệu quả chi phí hơn. 

- Kiến trúc ứng dụng serverless có thể phụ thuộc vào nhà cung cấp, có thể gây ràng buộc đối với khả năng của nền tảng.

**Vendor Lock-in (Ràng Buộc Nhà Cung Cấp)**

- Do kiến trúc serverless có khả năng mở rộng lên và giảm xuống phản hồi với công việc, đôi khi chúng cần bắt đầu từ không để phục vụ yêu cầu mới. 

- Đối với một số ứng dụng, độ trễ này không ảnh hưởng nhiều, nhưng đối với ứng dụng tài chính có độ trễ thấp, độ trễ này có thể không chấp nhận được.

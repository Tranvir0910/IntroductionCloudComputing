### **Microservices Infrastructure**

- Kiến trúc Microservices là **một phương pháp trong đó một ứng dụng lớn được tạo ra từ nhiều thành phần hoặc dịch vụ nhỏ**, được **kết nối linh hoạt và có thể triển khai độc lập**. 

- Những dịch vụ này **thường có các stack riêng chạy trên các container của chúng**. Chúng ***giao tiếp với nhau thông qua sự kết hợp của APIs, event streaming và message brokers***. 

- Đối với doanh nghiệp, điều này có nghĩa là **các thành phần ứng dụng có thể được phát triển** và **cập nhật một cách hiệu quả hơn bởi nhiều nhà phát triển làm việc độc lập**. 

- Các đội có thể **sử dụng các stack và môi trường chạy khác nhau cho các thành phần khác nhau**. 

- Các thành phần đối mặt với tải lớn có thể **được mở rộng độc lập**, **giảm lãng phí và chi phí liên quan đến việc phải mở rộng toàn bộ ứng dụng**.

#### **1. Thay Đổi Trong Phát Triển Ứng Dụng**

- Trong quá khứ, phần mềm được xây dựng dưới dạng các ứng dụng monolithic lớn, nơi một nhóm phát triển mất tháng để xây dựng một ứng dụng lớn dựa trên một mã nguồn chung. 

- Người phát triển viết từng phần của ứng dụng từ đầu đến cuối. 

- Tuy nhiên, sau hàng thập kỷ phát triển phần mềm, có rất nhiều mã nguồn đã được phát triển có sẵn mà nhà phát triển có thể sử dụng như là cơ sở của một ứng dụng, nghĩa là họ không cần phải tạo mọi dòng mã từ đầu. 

- Các nền tảng phát triển đám mây cung cấp cho nhà phát triển một hệ sinh thái mã nguồn mà họ có thể dễ dàng và an toàn tích hợp vào ứng dụng. 

- Thay vào đó, **thay vì xây dựng một ứng dụng lớn trên một nhóm**, **nhà phát triển chia nhỏ thành các nhóm nhỏ độc lập**, **nơi họ viết mã nguồn ít hơn được gọi là microservices**. 

- **Microservices chia nhỏ ứng dụng lớn thành các chức năng cốt lõi**, chẳng hạn như **Tìm kiếm, Đề xuất, Xếp hạng của Khách hàng, hoặc Danh mục Sản phẩm**.


#### **2. Trường Hợp Sử Dụng Microservices: Dream Game**

- Hãy xem cách Microservices hoạch định trải nghiệm người dùng trong một trường hợp sử dụng cụ thể. 

- Ron là một người hâm mộ bóng đá sử dụng dịch vụ truyền hình trực tuyến có tên Dream Game. 

- Hôm qua, anh ấy bỏ lỡ cơ hội xem trận bán kết quan trọng của đội mình. May mắn thay, anh ấy có thể xem lại trận đấu vào tối nay trên Dream Game. 

- Khi anh ấy đăng nhập, anh ấy thấy nội dung phổ biến nhất giữa tất cả người dùng Dream Game. 

- Sau một thời gian tìm kiếm, anh ấy tìm thấy trận đấu mà anh ấy đang tìm kiếm. Những gì anh ấy thực sự muốn là tìm trận đấu của mình chỉ bằng một cú nhấp chuột. 

- May mắn thay, đội phát triển Dream Game đang sử dụng Microservices để phát triển trải nghiệm người xem tốt hơn cho những người như Ron.

- **Microservice 1: Danh mục Nội dung**

    - Nhóm nhỏ nhà phát triển tổ chức từng phần nội dung với siêu dữ liệu mô tả chúng. 
    
    - **Dữ liệu này đưa vào Microservice 2: Chức năng Tìm kiếm**.

- **Microservice 2: Chức năng Tìm kiếm**

    - Chức năng Tìm kiếm đảm bảo rằng kết quả tìm kiếm của Ron được thu thập và so sánh với danh mục Dream Game. 
    
    - Dữ liệu này sẽ được **sử dụng để tạo ra kết quả tìm kiếm chính xác**.

- **Microservice 3: Đề Xuất**

    - Microservice này thu thập dữ liệu về nội dung phổ biến nhất giữa tất cả người dùng Dream Game. 
    
    - Đây là nội dung tạo nên trang chủ mà Ron thấy khi anh ấy đăng nhập.



#### 3. **Microservices Hoạt động như thể nào**

- Các Microservices này đều **được đặt trong các container riêng biệt, sẵn sàng để tham gia vào ứng dụng**. 

- **Trước khi có thể làm việc cùng nhau, chúng phải tìm thấy nhau thông qua Service Discovery**, **tạo ra một bản đồ để các Microservices có thể giao tiếp**. 

- Khi **chúng tìm thấy nhau, chúng giao tiếp sử dụng API**. 

- Khi Ron **tìm kiếm đội bóng đá yêu thích** của mình, **Microservice Tìm kiếm giao tiếp với Danh mục Nội dung (qua API) về điều Ron đang tìm kiếm**.

- Quay trở lại mục tiêu chính: **đưa Ron đến trận đấu bóng đá của mình chỉ với một cú nhấp chuột**. 

    - Nhóm phát triển làm việc trên Microservice Đề Xuất, cập nhật mã nguồn, **thêm vào một thuật toán phân tích**. 
    
    - Sử dụng phân tích, **Microservice Đề Xuất sẽ so sánh lịch sử xem của Ron và sở thích với nội dung phổ biến giữa người xem khác**, bao gồm người hâm mộ bóng đá và người xem trong khu vực địa lý và đối tượng như Ron. 
    
    - Vì nhà phát triển **không cần tạo mã từ đầu, họ có thể triển khai chức năng mới này trong vài ngày**. Những c**ập nhật này xảy ra ngầm trong khi các container Microservice còn lại hoạt động bình thường.**
    
    - Lần tiếp theo Ron kiểm tra Dream Game, **thay vì chỉ thấy nội dung phổ biến nhất hoặc mới nhất**, **anh ấy thấy một danh sách phát cá nhân sẽ tiếp tục làm rõ ràng theo thời gian khi hệ thống học thêm về thói quen và sở thích xem của anh ấy**. 
    
**Kết quả:** Ron tìm thấy trận đấu mới nhất của đội bóng yêu thích ngay lập tức. 

- Phương pháp Microservices **giúp nhà phát triển nhanh chóng đổi mới các ứng dụng theo cách song song và giúp người dùng như Ron tập trung vào những thứ thực sự quan tâm**. 

- Và khi những quan tâm đó thay đổi và phát triển nhanh chóng mỗi ngày, **Microservices giúp doanh nghiệp theo kịp và phát triển cùng khách hàng của mình**.

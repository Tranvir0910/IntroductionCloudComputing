### **Definition and Essential Characteristics of Cloud Computing**

- **Cloud computing** là một mô hình để cung cấp quyền truy cập mạng thuận tiện, theo yêu cầu, đến một nguồn tài nguyên máy tính có thể được cấu hình và cung cấp nhanh chóng với ít công sức quản lý hoặc tương tác với nhà cung cấp dịch vụ. 

- Các nguồn tài nguyên máy tính bao gồm mạng, máy chủ, lưu trữ, ứng dụng và dịch vụ.

- Mô hình đám mây này bao gồm năm đặc điểm cơ bản, bốn mô hình triển khai và ba mô hình dịch vụ.

#### **Five Essential Characteristics**

**Demand self service, Broad network access, Resource pooling, Rapid elasticity và Measured service.**

##### **On-demand self-service**

- Bạn có thể truy cập tài nguyên đám mây bất cứ khi nào cần.

- Dịch vụ luôn sẵn có trong suốt cả năm, không phụ thuộc vào việc một ngày cụ thể có phải là ngày lễ quốc gia, cuối tuần, ngày lễ và cetera. 

- Không có ngoại trừ trường hợp dịch vụ bị gián đoạn hoặc có sự vi phạm an ninh.

##### **Broad network access**

- Tài nguyên của đám mây có thể được truy cập thông qua mạng. 

- Dịch vụ đám mây công cộng có thể được truy cập từ bất kỳ đâu và trên bất kỳ thiết bị nào có kết nối Internet và khả năng duyệt web.

- Tuy nhiên, trong mô hình đám mây riêng tư tại chỗ, một mạng nội bộ có thể đủ để truy cập dịch vụ đám mây, miễn là tổ chức không cần phải liên lạc với toàn thế giới công cộng.

##### **Resource Pooling**

- Người tiêu dùng tiết kiệm chi phí khi sử dụng mô hình chia sẻ. Mô hình này cung cấp cho nhà cung cấp đám mây quy mô kinh tế, mà họ có thể truyền đạt cho khách hàng của họ. 

- Cung cấp một chi phí hiệu quả cho đám mây. Tài nguyên máy tính được gom lại để phục vụ nhiều người tiêu dùng bằng mô hình đa người sử dụng, tài nguyên đám mây được phân chia và phân chia lại theo yêu cầu mà không cần khách hàng quan tâm đến vị trí vật lý của những tài nguyên này.

##### **Rapid Elasticity**

- Bạn có thể tăng hoặc giảm tài nguyên theo yêu cầu của bạn. Do tính linh hoạt của đám mây, tài nguyên có thể được mở rộng hoặc thu nhỏ theo chiều dọc hoặc chiều ngang. 

- Một ví dụ lý tưởng là thêm tài nguyên khi có sự tăng cầu do số lượng người dùng tăng khi mua sắm trực tuyến, chẳng hạn như trong một chương trình giảm giá lễ hội, và ngược lại sau kỳ giảm giá.

##### **Measured Service**

- Bạn chỉ trả tiền cho những gì bạn sử dụng hoặc đặt trước theo yêu cầu. 

- Dịch vụ được đo lường, tuy nhiên, **không áp dụng** đối với một số dịch vụ đám mây như các **dịch vụ email chung** như **Gmail, Hotmail, Yahoo, các trang web truyền thông xã hội như Facebook, Twitter, WhatsApp và đánh giá dịch vụ đám mây thử nghiệm như AWS, Azure và GCP, được cung cấp miễn phí** do chính sách quyết định của các nhà cung cấp dịch vụ tương ứng. Dịch vụ được đo lường còn được gọi là mô hình tính cước, nơi bạn được tính phí sau khi sử dụng và vào cuối khoảng thời gian đã định, chẳng hạn như cước điện hàng tháng.

### **Lesson 1 Summary: Introduction to Cloud Computing**

- Điện toán đám mây là việc cung cấp các nguồn tài nguyên tính toán theo yêu cầu qua internet dựa trên hình thức trả tiền theo sử dụng; các nguồn tài nguyên được phân bổ và tái phân bổ động trong số nhiều người dùng và mở rộng và thu hẹp phù hợp với nhu cầu của người dùng.

- Nguyên gốc của điện toán đám mây có thể được truy ngược về các máy chủ chính của những năm 1950, với các công nghệ ảo hóa và hypervisors làm xúc tác cho sự xuất hiện của điện toán đám mây hiện đại.

- Các tổ chức phải xem xét nhu cầu kinh doanh, khả năng đầu tư và khả năng rủi ro để tạo ra một chiến lược áp dụng đám mây mang lại những lợi ích mong muốn mà không gây ra sự gián đoạn kinh doanh và các vấn đề về bảo mật, tuân thủ quy định hoặc hiệu suất.

- Sự áp dụng đám mây đang phát triển nhanh hơn dự đoán. Điều khiển làn sóng công nghệ này là các nhà cung cấp dịch vụ đám mây với một loạt các dịch vụ từ Cơ sở hạ tầng, Nền tảng và Phần mềm. Một số nhà cung cấp Đám mây lớn của thời đại chúng ta bao gồm AWS, Alibaba Cloud, Google, IBM và Microsoft Azure.

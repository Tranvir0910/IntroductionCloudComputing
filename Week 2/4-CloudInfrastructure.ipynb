### **Overview of Cloud Infrastructure**


- Sau khi lựa chọn mô hình dịch vụ đám mây và loại đám mây được cung cấp bởi các nhà cung cấp, khách hàng cần lên kế hoạch về kiến trúc cơ sở hạ tầng. 

- Tầng cơ sở hạ tầng là nền tảng của đám mây. Tầng này bao gồm các tài nguyên vật lý được đặt tại Các **Khu vực (Regions), Khu (Zones) và Trung tâm Dữ liệu (Data Centers).**

- Một Khu vực đám mây là một khu vực địa lý hoặc vị trí nơi cơ sở hạ tầng của một nhà cung cấp đám mây được tập trung, và có thể có tên như **NA South hoặc US East.**

- Các Khu vực đám mây được cách ly với nhau để nếu một Khu vực bị ảnh hưởng bởi thảm họa tự nhiên như động đất, các hoạt động đám mây ở các Khu vực khác vẫn tiếp tục hoạt động.

#### **Availability Zones**

- Mỗi Khu vực đám mây có thể có nhiều Khu (hoặc Khu Khả dụng, viết tắt là AZ), đó thường là các Trung tâm Dữ liệu riêng biệt với nguồn điện, làm mát và tài nguyên mạng riêng của chúng.

- Các Khu (AZ) này có thể có tên như **DAL-09 hoặc us-east-1**. 

- Việc cách ly các Khu **cải thiện khả năng chịu lỗi tổng thể của đám mây, giảm độ trễ và tránh tạo ra một điểm thất bại chia sẻ duy nhất**. 

- Các Khu Khả dụng (và Trung tâm Dữ liệu bên trong chúng) kết nối với các Khu (AZ) và Khu vực khác, trung tâm dữ liệu riêng và Internet bằng cách sử dụng kết nối mạng có băng thông rất cao.

#### **Cloud Data Center**

- Một Trung tâm Dữ liệu đám mây là một phòng lớn hoặc một kho chứa cơ sở hạ tầng đám mây. 

- Các trung tâm dữ liệu này **chứa các pod và racks hoặc các container chuẩn hóa của tài nguyên máy tính như máy chủ, cũng như thiết bị lưu trữ và mạng - gần như mọi thứ mà một môi trường IT vật lý có**.

#### **Computing Resource**

- Nhà cung cấp đám mây cung cấp nhiều tùy chọn tính toán, bao gồm **máy ảo (Virtual Servers), máy chủ vật lý (Bare Metal Servers), và tài nguyên tính toán "Không Server" (Serverless)**.

- Hầu hết các máy chủ trong một trung tâm dữ liệu đám mây **chạy hypervisors để tạo ra máy ảo hoặc máy ảo (còn được gọi là VMs)**, đó là máy tính dựa trên phần mềm, dựa trên các công nghệ ảo hóa. 

- Các máy chủ khác **trong các rack** là **máy chủ vật lý không được ảo hóa**. Khách hàng **có thể triển khai máy ảo và máy chủ vật lý khi cần và chạy các công việc làm của họ trên chúng**. 

- Người dùng đám mây cũng **có thể chạy các công việc làm của họ trên tài nguyên tính toán "Không Server"**, đó là một lớp trừu tượng trên các máy ảo.

#### **Storage**

- Thông tin và dữ liệu **có thể bao gồm tệp, mã, tài liệu, hình ảnh, video, bản sao lưu, ảnh chụp và cơ sở dữ liệu và có thể được lưu trữ trong nhiều loại lưu trữ khác nhau trên Đám mây**. 

- Máy chủ vật lý và máy ảo được triển khai với lưu trữ mặc định trong ổ đĩa cục bộ.

#### **Networking**

**Mạng:**

1. **Cơ sở Hạ tầng Mạng:**

    - Trong một trung tâm dữ liệu đám mây, cơ sở hạ tầng mạng bao gồm các thành phần truyền thống như **bộ định tuyến và switch**. 

    - Tuy nhiên, **quan trọng hơn** đối với người dùng đám mây là sự hiện diện của các tùy chọn **Mạng Được Định nghĩa Bằng Phần Mềm (SDN). SDN cho phép ảo hóa hoặc cung cấp tài nguyên mạng theo cách có thể lập trình thông qua APIs, giúp dễ dàng triển khai, cấu hình và quản lý mạng trong đám mây.**

2. **Giao Diện Mạng ( Network Interface ) :**

    - Khi máy chủ được triển khai trong đám mây, cần **thiết lập các giao diện mạng, bao gồm giao diện mạng công cộng và riêng tư**. 
    
    - Giao diện mạng công cộng kết nối máy chủ với internet công cộng, trong khi giao diện mạng riêng tư cung cấp kết nối đến các nguồn tài nguyên đám mây khác và giúp bảo vệ chúng.

3. **Quản lý Địa chỉ IP và Subnet:**

    - Các giao diện mạng trong đám mây cần có địa chỉ IP và subnet được gán tự động hoặc được cấu hình. Quản lý địa chỉ IP và subnet là quan trọng để đảm bảo hoạt động mạng một cách hiệu quả.

4. **Bảo mật và Quản lý Truy Cập:**

    - Trong môi trường đám mây, cấu hình nào có thể truy cập tài nguyên của bạn và lưu lượng mạng nào được phép được thực hiện thông qua việc thiết lập Nhóm Bảo mật và Danh Sách Kiểm soát Truy cập (ACLs). Điều này là quan trọng để đảm bảo an toàn và cách ly tài nguyên trong đám mây.

5. **Mạng Ảo và Các Tùy Chọn An Toàn:**

    - Để tăng cường an ninh và cách ly tài nguyên trong đám mây, các nhà cung cấp đám mây thường cung cấp các khái niệm như **Mạng Cục bộ Ảo (VLANs), Mạng Riêng Ảo (VPCs), và Mạng Riêng Ảo (VPNs)**. Các thiết bị phần cứng như **tường lửa, cân bằng tải, cổng và bộ phân tích lưu lượng có thể được ảo hóa và cung cấp dưới dạng dịch vụ trong đám mây**.

6. **Mạng Phân phối Nội dung (CDN):**

    - Một khả năng mạng khác do các nhà cung cấp đám mây cung cấp là M**ạng Phân phối Nội dung hoặc CDN. CDN phân phối nội dung đến nhiều điểm trên thế giới, giúp người dùng truy cập nhanh hơn bằng cách lấy nó từ một điểm gần họ**.

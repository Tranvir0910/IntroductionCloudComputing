### **Secure Network in Cloud**

#### **Networking on Cloud vs. Premise**

- Khái niệm xây dựng mạng Đám mây không khác nhiều so với triển khai một mạng trong một trung tâm dữ liệu on-premises. 

- Sự khác biệt chính phát sinh từ việc trong Đám mây, chúng ta sử dụng các phiên bản logic của các yếu tố mạng thay vì các thiết bị vật lý. 

- Ví dụ, bộ điều khiển giao diện mạng hoặc NICs sẽ được đại diện bằng vNICs trong môi trường Đám mây. Trong Đám mây, các chức năng mạng được cung cấp dưới dạng dịch vụ thay vì dưới dạng các thiết bị đặt trên rack.

#### **Triển Khai Mạng Đám Mây**

- Để tạo ra một mạng trong Đám mây, người ta bắt đầu bằng cách **xác định kích thước của mạng hoặc dải địa chỉ IP thiết lập ranh giới cho mạng Đám mây**.

- Mạng Đám mây được triển khai trong các không gian mạng là các đoạn mạng được phân chia logic của các mạng sử dụng các tùy chọn bao gồm Mạng Ảo Riêng, hoặc VPC, có thể được chia thành các đoạn nhỏ hơn gọi là subnets.

- Các mạng Đám mây được phân đoạn logic là phần riêng tư của Đám mây **mang lại cho khách hàng sự bảo mật của các đám mây riêng và tính mở rộng của các đám mây công cộng**.

- Các nguồn lực Đám mây như **VMs hoặc các trường hợp máy chủ ảo, storage, kết nối mạng và cân bằng tải được triển khai vào các subnets**. Việc sử dụng subnets cho phép người dùng triển khai các ứng dụng doanh nghiệp bằng cách sử dụng các khái niệm đa tầng giống như trong môi trường on-premises. 

- Subnets cũng là khu vực chính nơi triển khai bảo mật trong Đám mây. 

- Mỗi subnet được bảo vệ bởi danh sách điều khiển truy cập hoặc ACLs, chúng hoạt động như một tường lửa cấp subnet.

- Bên trong subnet, người ta có thể tạo ra các nhóm an ninh cung cấp bảo mật ở cấp độ thể hiện, như các trường hợp máy ảo. 

- Khi bạn xây dựng một subnet, sau đó là thời điểm để thêm một số trường hợp máy ảo và bộ nhớ vào đó để bạn có thể chạy các ứng dụng của mình.

- Hãy tưởng tượng bạn có một **ứng dụng ba tầng yêu cầu truy cập Web VSIs**, các trường hợp máy ảo ở tầng ứng dụng và các trường hợp máy ảo cơ sở dữ liệu ở phía sau. 

- Trong trường hợp này, chúng ta sẽ đặt các **VSI truy cập web vào một nhóm an ninh**, các **VSI ứng dụng vào một nhóm an ninh thứ hai, trong khi các VSI cơ sở dữ liệu vào một nhóm an ninh thứ ba**. 

- Không cần phải nói, các **VSI truy cập Web cần có quyền truy cập Internet**. Một trường hợp **cổng Internet công cộng được thêm vào mạng để cho phép người dùng truy cập ứng dụng trong tầng Internet**.

- Trong khi cổng Internet công cộng là lựa chọn tuyệt vời cho việc truy cập Internet đến Đám mây, các doanh nghiệp quan tâm đến mở rộng tài nguyên on-premises của họ đến Đám mây bằng cách kết nối chúng một cách an toàn thông qua Mạng Riêng Ảo hoặc VPN.

- Khi xây dựng nhiều subnets và triển khai nhiều tải công việc, trở nên cần thiết để đảm bảo rằng các ứng dụng vẫn tiếp tục phản hồi. 

- Điều này được đạt được bằng cách sử dụng **cân bằng tải ( load balancer )** đảm bảo khả dụng băng thông cho các ứng dụng khác nhau.

#### **Direct Connectivity**

- Các doanh nghiệp với môi trường Đám mây hybrid thường sử dụng các kết nối tốc độ cao dành riêng giữa các Đám mây và tài nguyên on-premises là một cách an toàn và hiệu quả hơn so với các giải pháp kết nối công cộng. 

- Một số nhà cung cấp dịch vụ đám mây cung cấp kết nối như vậy, chẳng hạn như IBM Cloud và giải pháp Direct Link của nó, cho phép mở rộng tài nguyên on-premises đến Đám mây khi cần thiết.

#### **Building a Cloud Network**

- Xây dựng Mạng Đám mây đòi hỏi việc tạo ra một tập hợp các cấu trúc logic cung cấp chức năng mạng tương tự như các mạng trung tâm dữ liệu mà tất cả các chuyên gia IT đã đang dựa vào để bảo vệ môi trường và đảm bảo ứng dụng doanh nghiệp hoạt động hiệu quả.

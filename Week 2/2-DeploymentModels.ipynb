### **Public Cloud**

- Trong mô hình đám mây công cộng, người dùng có **quyền truy cập vào máy chủ, lưu trữ, mạng, bảo mật và ứng dụng dưới dạng dịch vụ được cung cấp bởi các nhà cung cấp dịch vụ đám mây qua internet**. 

- Sử dụng các bảng điều khiển web và API, người dùng có thể cấu hình tài nguyên và dịch vụ mà họ cần. 

- Nhà cung cấp đám mây sở hữu, quản lý, cấp phát và duy trì cơ sở hạ tầng, cho thuê nó cho khách hàng dưới dạng cước đăng ký hoặc phí sử dụng.

- Người dùng không sở hữu các máy chủ mà ứng dụng chạy trên đó hoặc lưu trữ dữ liệu mà họ tiêu thụ, cũng như không quản lý hoạt động của máy chủ hoặc xác định cách nền tảng được duy trì.

- Giống như cách chúng ta tiêu thụ và thanh toán cho các tiện ích như nước, điện, hoặc gas trong cuộc sống hàng ngày, chúng ta không sở hữu bất kỳ tài nguyên đám mây nào – chúng ta ký một thỏa thuận với nhà cung cấp dịch vụ, sử dụng tài nguyên và thanh toán cho việc sử dụng trong một khoảng thời gian nhất định.

- Đám mây công cộng mang lại sự tiết kiệm chi phí đáng kể về Tổng chi phí Sở hữu (TCO) vì nhà cung cấp chịu toàn bộ chi phí vốn, chi phí hoạt động và bảo dưỡng cho cơ sở hạ tầng và các cơ sở mà chúng được đặt. 

- Nó làm cho việc mở rộng trở nên dễ dàng như việc yêu cầu thêm dung lượng. 

- Tuy nhiên, với đám mây công cộng, người dùng không có bất kỳ kiểm soát nào đối với môi trường tính toán và phải tuân thủ hiệu suất và bảo mật của cơ sở hạ tầng của nhà cung cấp đám mây.

- Có nhiều nhà cung cấp đám mây công cộng trên thị trường ngày nay, như **Amazon Web Services, Microsoft Azure, IBM Cloud, Google Cloud Platform và Alibaba Cloud.** 

- Mặc dù tất cả các nhà cung cấp đều bao gồm một bộ dịch vụ cơ bản chung, như máy chủ, lưu trữ, mạng, bảo mật và cơ sở dữ liệu, họ cũng cung cấp một loạt các dịch vụ chuyên sâu với các tùy chọn thanh toán đa dạng.

**Một số đặc điểm của Public Cloud:**

- Đám mây công cộng là một kiến trúc đa người sử dụng ảo hóa cho phép người sử dụng chia sẻ tài nguyên máy tính, nằm ngoài tường lửa của họ. 

- Hồ bơi tài nguyên của nhà cung cấp đám mây, bao gồm cơ sở hạ tầng, nền tảng và phần mềm, KHÔNG được dành riêng cho việc sử dụng của một người sử dụng hoặc tổ chức duy nhất. 

- Tài nguyên được phân phối theo nhu cầu được cung cấp thông qua nhiều mô hình đăng ký và thanh toán theo giờ.

**Lợi ích của Public Cloud**

- Tài nguyên lớn có sẵn theo yêu cầu, cho phép ứng dụng phản ứng mượt mà với biến động trong nhu cầu. 

- Xem xét số lượng lớn người sử dụng chia sẻ tài nguyên đám mây trung ương theo yêu cầu, đám mây công cộng mang lại quy mô kinh tế lớn nhất. 

- Số lượng lớn các tài nguyên máy chủ và mạng có sẵn trên đám mây công cộng có nghĩa là đám mây công cộng có sẵn rất nhiều — nếu một thành phần vật lý thất bại, dịch vụ vẫn chạy bình thường trên các thành phần khác có sẵn.

**Một số lo ngại mà người dùng có đối với Public Cloud**

- An ninh và tuân thủ quy định về chủ quyền dữ liệu.

- Với dữ liệu được lưu trữ ở các địa điểm khác nhau và được truy cập qua biên giới quốc gia, việc tuân thủ các quy định về chủ quyền dữ liệu, quy định về chuyển giao và bảo mật dữ liệu đã trở nên ngày càng quan trọng đối với các doanh nghiệp. 

- Khả năng của nhà cung cấp dịch vụ không chỉ để duy trì các quy định, mà còn để hiểu và giải thích đúng các quy định này, là một vấn đề mà nhiều doanh nghiệp đều quan tâm.

**Trường hợp sử dụng Public Cloud**

- Tổ chức ngày càng ưa chuộng sử dụng đám mây để tập trung xây dựng ứng dụng, giảm thời gian đưa sản phẩm ra thị trường. 

- Doanh nghiệp nhu cầu linh hoạt thường chọn đám mây công cộng, sử dụng tính toán công cộng cho phục hồi sau thảm họa và bảo vệ dữ liệu. 

- Nhiều tổ chức chuyển sang lưu trữ và quản lý dữ liệu đám mây để cải thiện khả năng truy cập và sao lưu. 

- Bộ phận Công nghệ thông tin thường outsourced quản lý nền tảng và ứng dụng ít quan trọng đến nhà cung cấp đám mây công cộng.

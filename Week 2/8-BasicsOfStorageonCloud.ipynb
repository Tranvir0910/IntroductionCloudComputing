### **Basics of Storage on Cloud**

#### **Introduction to Cloud Storage:**

- Lưu trữ đám mây ( Cloud Storage ) là nơi bạn lưu trữ các tệp dữ liệu trong đám mây. Một số lưu trữ phải được kết nối với một nút tính toán trước khi có thể truy cập, trong khi loại lưu trữ khác có thể được truy cập trực tiếp thông qua Internet công cộng hoặc kết nối mạng riêng tư.

- Nhà cung cấp dịch vụ đám mây **đảm bảo lưu trữ và cơ sở hạ tầng liên quan được lưu trữ, bảo vệ, quản lý và duy trì** để đảm bảo bạn có quyền truy cập dữ liệu khi cần. 

- Dịch vụ lưu trữ đám mây cho phép bạn **mở rộng dung lượng theo nhu cầu để bạn chỉ trả tiền cho những gì bạn cung cấp, thường là dựa trên cơ sở 'mỗi gigabyte'**. 

- Chi phí lưu trữ sẽ thay đổi theo loại, nhưng nói chung, **đọc/ghi nhanh hơn của bộ nhớ, chi phí trên mỗi gigabyte càng cao**. 

- Lưu trữ đám mây có bốn loại chính - **Direct Attached, File Storage, Block Storage và Object Storage**.

#### **Direct Attached Storage**

- Direct Attached Storage, đôi khi được gọi là **'Lưu trữ Local'**, là lưu trữ được trình bày trực tiếp cho một máy chủ dựa trên đám mây và thực sự nằm trong thân máy chủ hoặc cùng một rack.

- Lưu trữ này **nhanh chóng và thường chỉ được sử dụng để lưu trữ hệ điều hành của máy chủ, mặc dù nó cũng có thể có các ứng dụng sử dụng khác**.

- Hai lý do chính tại sao lưu trữ được đính kèm trực tiếp không phải là lựa chọn tốt cho các ứng dụng khác ngoài việc lưu trữ hệ điều hành là: 
    
    - Nó thường **chỉ tồn tại trong thời gian tài nguyên tính toán nó được đính kèm**.
    
    - Nó **không thể được chia sẻ với các nút khác và mặc dù bạn có thể sử dụng kỹ thuật RAID, nhưng nó không đủ chống lại lỗi như các loại lưu trữ khác**.

#### **File Storage**

- Lưu trữ tệp thường được trình bày cho các nút tính toán dưới dạng **'Lưu trữ NFS'**. 

- **NFS** có nghĩa là **Hệ thống Tệp Mạng** và có nghĩa là **lưu trữ được kết nối với các nút tính toán qua một mạng Ethernet tiêu chuẩn**. 

- Lưu trữ được gắn kết qua NFS phổ biến, nhưng **thường chậm hơn so với lưu trữ được gắn kết trực tiếp hoặc lưu trữ khối vì dữ liệu đi qua mạng Ethernet**. Nó cũng **thường rẻ hơn so với lưu trữ được gắn kết trực tiếp hoặc lưu trữ khối**.

#### **Block Storage**

- Lưu trữ khối được **trình bày cho các nút tính toán bằng cách sử dụng kết nối sợi quang tốc độ cao**, có nghĩa là **tốc độ đọc/ghi thường nhanh** và **đáng tin cậy hơn so với lưu trữ tệp**, làm cho lưu trữ khối phù hợp để sử dụng với cơ sở dữ liệu và các ứng dụng khác nơi tốc độ đĩa quan trọng. 

- Thông thường, bạn sẽ **cung cấp lưu trữ khối theo 'thể tích'**, sau đó có thể **được gắn kết vào một nút tính toán**, mà nó sau đó **hiệu quả nhìn thấy như một ổ cứng khác**. Thông thường, thì **chỉ có thể gắn kết 'thể tích' vào một nút tính toán một lúc**.

- **IOPS là viết tắt của 'Input/Output Operations Per Second'** và **liên quan đến tốc độ của lưu trữ, hoặc nói cách khác, cách nhanh chóng dữ liệu có thể được đọc từ hoặc ghi vào lưu trữ**. 

- **'Persistence'** là **một thuật ngữ được sử dụng khi triển khai lưu trữ Tệp hoặc Khối** và liên quan đến điều gì xảy ra với lưu trữ sau khi nút tính toán mà nó đính kèm đã bị chấm dứt.

- Nếu lưu trữ được đặt là **'persist'**, thì nó sẽ **không bị xóa cùng với nút tính toán**, có nghĩa là **nó và dữ liệu của nó được bảo tồn và có sẵn để gắn kết vào một nút tính toán khác, tuy nhiên, bạn sẽ tiếp tục trả tiền cho lưu trữ đó**. 

- Bạn cũng có thể, trong một số trường hợp, **đặt lưu trữ sao cho nó tự động bị xóa khi nút tính toán mà nó được gắn kết vào bị xóa** - trong trường hợp này, như chúng ta biết, **nó trở thành 'Ephemeral Storage'**. 

- Ở đây, bạn cũng sẽ **ngừng trả tiền cho lưu trữ nhưng bạn sẽ mất mọi dữ liệu trừ khi nó được sao lưu ở một nơi nào đó**.

- Có nhiều cách để sao lưu dữ liệu trong đám mây, nhưng **một cách để sao lưu cả lưu trữ Tệp lẫn Khối là tạo một 'Snapshot'** là **hình ảnh của lưu trữ**, **tại một thời điểm cụ thể**. 

- **Snapshot thường nhanh chóng để tạo** (họ thực sự không ghi bất kỳ dữ liệu nào, hoặc nói cách khác, họ tạo ra siêu dữ liệu), **không đòi hỏi thời gian tạm nghỉ và các snapshot sau này chỉ ghi lại các thay đổi vào dữ liệu**. 

- Chúng rất tốt để **trả lại lưu trữ về trạng thái nó ở một Snapshot cụ thể**, mặc dù lưu ý, **chúng không thể được sử dụng để khôi phục các tệp tin cá nhân**.

#### **Object Storage**

- Loại lưu trữ thứ tư là Lưu trữ Đối tượng. Đây là một loại lưu trữ khác biệt vì **nó không được đính kèm vào một nút tính toán, thay vào đó, nó được truy cập thông qua một API**. 

- Trong tất cả các loại lưu trữ, **Lưu trữ Đối tượng là rẻ nhất và cũng chậm nhất về tốc độ đọc/ghi, nhưng nó vô hạn với người dùng cuối**.  

- Khác với Lưu trữ Tệp và Lưu trữ Khối nơi bạn cung cấp một dung lượng lưu trữ nhất định và nó đầy lên theo thời gian, với Lưu trữ Đối tượng, bạn có thể tiếp tục thêm các tệp tin vào nó và nó không bao giờ đầy, **bạn chỉ trả tiền cho những gì bạn sử dụng**.

- Điều này khiến cho Lưu trữ Đối tượng trở thành một **kho chứa tuyệt vời cho mọi loại dữ liệu không cấu trúc, lớn và nhỏ, bao gồm tài liệu, video, log, sao lưu, dữ liệu từ IoT, ảnh ảo và hình ảnh máy ảo**.

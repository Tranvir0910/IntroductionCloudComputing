### **Object Storage**

#### **OS can be used without connecting to particular compute node to use it**

**Không cần kết nối nó với một nút tính toán cụ thể để sử dụng nó**

- Cấu hình một trường hợp dịch vụ lưu trữ đối tượng và **sử dụng một API hoặc giao diện lập trình ứng dụng để tải lên, tải xuống và quản lý dữ liệu của bạn**. 

- Điều này có nghĩa là bạn **có thể trực tiếp sử dụng lưu trữ đối tượng với bất cứ thứ gì bạn có thể gọi là API mà không cần một nút tính toán nền**.

#### **OS is less expensive than other cloud storage options**

**Ít tốn kém hơn so với các lựa chọn lưu trữ đám mây khác.**

Chi phí mỗi gigabyte thường chỉ là vài xu Mỹ mỗi tháng, và đôi khi thậm chí ít hơn, phụ thuộc vào loại lưu trữ được sử dụng - sẽ nói thêm về các cấp độ lưu trữ sau.

#### **The most important thing to note about OS is that it's effectively infinite**

**Điều quan trọng nhất về lưu trữ đối tượng là nó hiệu quả về mặt không gian lưu trữ**

- Bạn **chỉ định kích thước lưu trữ bạn muốn theo đơn vị gigabyte hoặc terabyte, rồi trả phí dựa trên kích thước bạn đã cấu hình.** 

- Với lưu trữ đối tượng, **bạn chỉ tiêu thụ không gian lưu trữ bạn cần và trả chi phí dựa trên chi phí cho mỗi gigabyte bạn sử dụng**. 

- Bạn có thể **tiếp tục tải lên các tệp và không gian lưu trữ sẽ không bao giờ cạn kiệt**.

#### **Object Storage Buckets**

- Lưu trữ đối tượng **tuyệt vời cho việc lưu trữ lượng lớn dữ liệu không có cấu trúc**. 

- Dữ liệu không có cấu trúc **có nghĩa là dữ liệu không được lưu trữ theo bất kỳ cấu trúc thư mục hay thư mục nào**.

- Lưu trữ đối tượng **sử dụng "buckets" (thùng) và các đối tượng được lưu trữ trong những "buckets" này theo cách cấu trúc phẳng**. 

- Một "bucket" **giống như một thư mục**, **có thể đặt tên có ý nghĩa và, tất nhiên, có thể có các "buckets" khác nhau cho các loại đối tượng khác nhau**. 

- Tuy nhiên, bạn **không thể đặt một "bucket" bên trong một "bucket"**. Khi một **đối tượng được đặt trong một "bucket", nó cũng có một số dữ liệu về dữ liệu được thêm vào, chẳng hạn như một ID đối tượng**. 

- Dữ liệu này **giúp ứng dụng định vị và truy cập đối tượng**, cũng như **cung cấp thông tin về thời gian dữ liệu được lưu trữ hoặc truy cập lần cuối**.

- Khi bạn tạo một "bucket", bạn **không cần cung cấp hoặc xác định bất kỳ thông tin kích thước nào**. 

- "Bucket" sẽ **chỉ giữ dữ liệu bạn đặt vào bên trong, và nhà cung cấp dịch vụ đảm bảo có đủ dung lượng lưu trữ**. 

- "Buckets" có thể **giữ ít nhất là vài byte dữ liệu, lên đến nhiều petabyte, và bạn có thể xây dựng lên lượng dữ liệu lưu trữ một cách từ từ hoặc nhanh chóng, cũng như giảm nó lại**.

#### **Managed by Service provider**

- Nhà cung cấp dịch vụ cũng đảm bảo tính phục hồi và đảm bảo rằng giải pháp lưu trữ đối tượng là **có sẵn với mức độ cao**. 

- Một số nhà cung cấp đám mây **cung cấp các loại "buckets" khác nhau với mức độ phục hồi khác nhau**. 

- Ví dụ, họ có thể cung cấp các "buckets" có độ bền, nhưng dữ liệu chỉ được lưu trữ ở một trung tâm dữ liệu. Đây là một lựa chọn tốt khi dữ liệu cần nằm ở một địa điểm địa lý cụ thể hoặc trong các tình huống nơi tính sẵn có cao hơn là một vấn đề. 

- Sau đó, họ sẽ cung cấp các "buckets" có tính sẵn có cao ở nhiều khu vực, nơi dữ liệu được lưu trữ nhiều lần ở các trung tâm dữ liệu hoặc vùng khác nhau trong cùng một khu vực hoặc thậm chí ở nhiều khu vực. 

- Các tùy chọn này thường có chi phí cao hơn, nhưng cung cấp cả mức độ độ bền và sẵn có cao nhất cho dữ liệu của bạn. 

- Dữ liệu này có thể là bất kỳ thứ gì, từ **tệp văn bản đến tệp âm thanh và tệp video, từ dữ liệu IoT đến hình ảnh máy ảo, từ tệp sao lưu đến lưu trữ dữ liệu lâu dài.** 

- **Đối với bất kỳ dữ liệu nào là tĩnh và không đòi hỏi tốc độ đọc và ghi nhanh, lưu trữ đối tượng sẽ là lựa chọn phù hợp.** 

- Tuy nhiên, lưu trữ đối tượng sẽ **không phù hợp để chạy các hệ điều hành hoặc các ứng dụng khác như cơ sở dữ liệu hoặc bất kỳ điều gì khác nơi nội dung của các tệp thay đổi**.

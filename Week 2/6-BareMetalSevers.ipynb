### **Bare Metal Severs**

#### **What is Bare Metal Severs**

- Một máy chủ bare metal là **một máy chủ vật lý chuyên dụng dành cho một khách hàng duy nhất**. 

- Nó được **dành riêng cho một khách hàng duy nhất**. Nhà cung cấp đám mây **thực sự lấy máy chủ vật lý và kết nối nó vào một tủ trong một trung tâm dữ liệu để cung cấp cho khách hàng**. 

- Nhà cung cấp đám mây **quản lý máy chủ đến hệ điều hành hoặc OS**, điều này có nghĩa là **nếu có vấn đề nào với phần cứng hoặc kết nối tủ**, họ sẽ **sửa chữa hoặc thay thế** nó, và **sau đó khởi động lại máy chủ**. **Khách hàng chịu trách nhiệm quản lý và quản trị mọi thứ khác trên máy chủ**.

#### **Bare metal server configuration**

- Máy chủ bare metal có thể được **cấu hình trước bởi nhà cung cấp đám mây** để **đáp ứng các gói công việc hoặc chúng có thể được cấu hình tùy chỉnh theo đặc tả của khách hàng**. 

- Điều này bao gồm **bộ xử lý, RAM, ổ cứng, các thành phần chuyên dụng và hệ điều hành**. Khách hàng cũng **có thể cài đặt hệ điều hành của riêng họ và có thể cài đặt một số hypervisors không có sẵn từ nhà cung cấp đám mây, từ đó tạo ra các máy ảo và farm riêng của họ**.

- Với máy chủ bare metal, bạn cũng có thể thêm GPU, được thiết kế để **tăng tốc tính toán khoa học, phân tích dữ liệu và hiển thị đồ họa ảo chuyên nghiệp**. **Do máy chủ bare metal là máy vật lý, nó mất thời gian lâu hơn để cấu hình so với máy chủ ảo**. 

- Các cấu hình trước của máy chủ bare metal có thể mất từ 20-40 phút để triển khai và cấu hình tùy chỉnh có thể mất khoảng ba hoặc bốn giờ. Nhưng thời gian cấu hình này có thể thay đổi tùy thuộc vào nhà cung cấp đám mây.

#### **Characteristics ( Đặc trưng )**

- Vì máy chủ bare metal được dành riêng cho việc sử dụng bởi một khách hàng duy nhất tại bất kỳ thời điểm nào, chúng thường có giá cao hơn so với máy ảo có cùng kích thước. 

- Cũng lưu ý rằng, khác với máy chủ ảo, không phải tất cả các nhà cung cấp đám mây đều cung cấp máy chủ bare metal. 

- Do máy chủ bare metal có thể tùy chỉnh hoàn toàn, chúng có thể thực hiện những gì khách hàng muốn trong môi trường đòi hỏi cao nhất.

#### **Workloads**

- Máy chủ bare metal được thiết kế và dành cho việc sử dụng hiệu suất cao và lâu dài trong môi trường an ninh và cách ly cao cấp. **Khách hàng có đầy đủ quyền truy cập và kiểm soát máy chủ bare metal vì không cần hypervisor**. 

- Do không chia sẻ phần cứng máy chủ cơ bản với các khách hàng khác, máy chủ bare metal đáp ứng nhu cầu đòi hỏi của **tính toán hiệu suất cao hoặc HPC và ứng dụng yêu cầu độ trễ tối thiểu**. 

- Những máy chủ này cũng xuất sắc trong các ứng dụng phân tích dữ liệu lớn và giải pháp yêu cầu GPU. 

- Một số ví dụ về công việc mà máy chủ bare metal **đáp ứng bao gồm ERP, CRM, AI, deep learning và ảo hóa**. 

- Nếu bạn sử dụng bất kỳ ứng dụng nào đòi hỏi mức kiểm soát bảo mật cao hoặc các ứng dụng mà bạn thường chạy trong môi trường on-premises, thì máy chủ bare metal là một lựa chọn tốt.

- Khi so sánh máy chủ bare metal với máy chủ ảo, một số yếu tố quan trọng cần xem xét nằm trong nhu cầu của khách hàng. 

- Máy chủ bare metal hoạt động tốt nhất cho các công việc **đòi hỏi nhiều CPU và I/O**, **xuất sắc trong hiệu suất và bảo mật, đáp ứng các yêu cầu tuân thủ nghiêm ngặt, cung cấp sự linh hoạt, kiểm soát và tính minh bạch hoàn chỉnh, nhưng đi kèm với chi phí quản lý và vận hành bổ sung**. 

- Ngược lại, **máy chủ ảo được triển khai nhanh chóng, cung cấp môi trường linh hoạt và có khả năng mở rộng, và chi phí sử dụng thấp**. Tuy nhiên, vì chúng **chia sẻ phần cứng cơ bản** với các máy chủ ảo khác, chúng **có thể bị giới hạn về thông lượng và hiệu suất**.

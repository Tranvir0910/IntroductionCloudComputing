### **VMs Virtual Machines**

- **Virtual Machines** hoặc VMs còn được biết đến là **Virtual Servers** hoặc **Virtual Instances**, hoặc **đơn giản là instances**, tùy thuộc vào nhà cung cấp đám mây. 

- Các nhà cung cấp đám mây khác nhau đều cung cấp VMs trong nhiều cấu hình và tùy chọn triển khai để phục vụ các trường hợp sử dụng khác nhau. 

- Khi bạn tạo một máy chủ ảo trong đám mây, bạn chỉ định Khu vực và Khu hoặc Trung tâm Dữ liệu bạn muốn máy chủ được triển khai và Hệ điều hành bạn muốn cài đặt.

- Bạn có thể chọn giữa VMs được chia sẻ (nghĩa là, đa người sử dụng) hoặc VMs được dành riêng (nghĩa là, một người sử dụng) . 

- Bạn cũng có thể chọn giữa thanh toán theo giờ hoặc hàng tháng và chọn tùy chọn lưu trữ và mạng cho máy chủ ảo. Bây giờ hãy xem xét một số loại VMs khác nhau có thể được triển khai trong đám mây.

#### **Shared hoặc Public Cloud VMs ( Máy ảo được chia sẻ hoặc đám mây công cộng ):**

- Được quản lý bởi nhà cung cấp, đây là triển khai đa người sử dụng có thể được triển khai theo yêu cầu với các kích thước được định trước.

- Giá cả và cấu hình được đề xuất cho các công việc tính toán khác nhau, từ công việc tốn nhiều tài nguyên CPU và RAM đến công việc tốn nhiều bộ nhớ.

- Có cả các tùy chọn tùy chỉnh cho phép người dùng xác định số lượng core, RAM và đặc tính lưu trữ.

#### **Transient hoặc Spot VMs ( Máy ảo tạm thời hoặc máy ảo Spot ):**

- Tận dụng khả năng không sử dụng trong trung tâm dữ liệu đám mây, giá cả thấp hơn so với VMs thông thường.

- Tuy nhiên, có thể bị hủy bỏ bất cứ lúc nào để cung cấp tài nguyên cho VMs thông thường với giá cao hơn.

- Thích hợp cho công việc không sản xuất như kiểm thử và phát triển ứng dụng.

#### **Reserved Virtual Server Instances ( Phiên bản máy chủ ảo dự trữ ):**

- Cho phép bạn đặt trước dung lượng và đảm bảo tài nguyên cho triển khai tương lai.

- Bạn có thể chọn một mức giá thấp hơn bằng cách cam kết sử dụng trong thời gian dài, chẳng hạn 1 năm hoặc 3 năm.

- Được đảm bảo dung lượng này trong trung tâm dữ liệu bạn chọn trong suốt thời gian hợp đồng.

#### **Dedicated Hosts ( Máy chủ chuyên dụng ):**

- Cung cấp cách ly đơn người sử dụng, chỉ có VMs của bạn chạy trên một máy chủ nhất định.

- Bạn cần chỉ định trung tâm dữ liệu và POD khi triển khai một máy chủ được dành riêng.

- Thường được sử dụng để đáp ứng các yêu cầu về tuân thủ và quy định hoặc đáp ứng các điều khoản cấp phép cụ thể.

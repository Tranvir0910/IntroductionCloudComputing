### **File Storage**

- Giống như lưu trữ đính kèm trực tiếp, lưu trữ tệp phải được đính kèm vào một nút tính toán trước khi có thể truy cập và có thể lưu trữ dữ liệu trên đó. 

- Tuy nhiên, Lưu trữ Tệp có thể **ít tốn kém hơn**, **chống lại lỗi tốt** hơn và **liên quan đến ít quản lý và bảo dưỡng ổ đĩa hơn so với lưu trữ đính kèm trực tiếp**. 

- Bạn cũng có thể **cung cấp một lượng lớn lưu trữ tệp** và **trình bày nó như một ổ đĩa cho một máy chủ**. 

- Lưu trữ tệp được **gắn kết từ các thiết bị lưu trữ từ xa**. Đó là, **các đĩa vật lý được chứa trong một thiết bị phần cứng chuyên dụng riêng biệt** và **sau đó được kết nối với nút tính toán qua cơ sở hạ tầng cơ bản trong trung tâm dữ liệu**. 

- Những thiết bị lưu trữ này **không chỉ cực kỳ chống lại lỗi**, mà **dữ liệu cũng an toàn hơn trong chúng vì những thiết bị lưu trữ này cung cấp các dịch vụ như mã hóa trong quá trình truyền**.

#### **Ethernet network**

- Những thiết bị này được quản lý bởi nhà cung cấp dịch vụ. **Lưu trữ Tệp được gắn kết vào nút tính toán thông qua mạng Ethernet - cùng loại mạng mà bạn có thể nhận email hoặc duyệt Internet qua, mặc dù mạng Ethernet này thường được dành riêng cho công việc này**. 

- Điều này có nghĩa là **đôi khi nó có thể được gọi là 'Lưu trữ Đính Kèm Mạng', 'Lưu trữ Tệp Mạng' hoặc đơn giản là 'NFS'**. Một trong những vấn đề với mạng Ethernet là **tốc độ của nó có thể thay đổi - mạng Ethernet càng tải trọng, khả năng tốc độ hoặc băng thông của nó bị ảnh hưởng càng cao**. 

- Tất nhiên, Nhà cung cấp Đám mây xây dựng mạng lưu trữ của họ để xử lý lưu lượng rất lớn, nhưng thậm chí vậy, tốc độ nhất quán không thể đảm bảo. Do đó, lưu trữ Tệp **thường được sử dụng cho các khối công việc nơi tốc độ mạng nhất quán cao không phải là một yêu cầu**.

#### **Multiple Compute Nodes**

- Về khối công việc, Lưu trữ Tệp thường **có thể được gắn kết vào nhiều nút tính toán cùng một lúc**, **trong đó ổ đĩa hoặc thể tích đã gắn kết chỉ giống như một ổ đĩa khác trên nút tính toán**. 

- **Khả năng gắn kết Lưu trữ Tệp vào nhiều nút tính toán cùng một lúc** làm cho nó **trở thành một giải pháp lý tưởng khi một số lưu trữ chung là cần thiết** - ví dụ, một thư mục chia sẻ của bộ phận, một 'khu vực đổ' cho các tệp tin đang nhập cần được xử lý bởi một ứng dụng, hoặc một kho lưu trữ các tệp tin mà một dịch vụ web có thể truy cập. 

- Trong những ứng dụng này, **sự biến động tiềm ẩn trong tốc độ của mạng kết nối thực sự không phải là một vấn đề**. Tất nhiên, **khi chi phí là một vấn đề**, bạn có thể **sử dụng lưu trữ tệp cho các ứng dụng khác như cơ sở dữ liệu, nhưng sự đánh đổi là tốc độ**. 

#### **IOPS**

- Khi bạn cung cấp lưu trữ tệp, một trong những điều cần xem xét là khả năng IOPS của lưu trữ. 

- **IOPS viết tắt của Input/Output Operations Per Second** và đề cập đến tốc độ mà các đĩa có thể ghi và đọc dữ liệu (lưu ý đây không phải là tốc độ của mạng giữa lưu trữ và nút tính toán). 

- **Giá trị IOPS càng cao, tốc độ của ổ đĩa cơ sở càng nhanh.** Một giá trị IOPS cao cũng **thường làm tăng chi phí. Hiểu biết về IOPS là quan trọng vì nếu giá trị IOPS quá thấp cho ứng dụng của bạn, lưu trữ có thể trở thành một chướng ngại và làm cho ứng dụng chạy chậm.** 

- Ngược lại, **nếu IOPS quá cao, bạn có thể phải trả nhiều hơn những gì bạn cần cho lưu trữ của mình**.

- Ví dụ, một thư mục chia sẻ có thể được gắn kết vào 30 nút tính toán khác nhau và một ứng dụng ghi và yêu cầu dữ liệu từ thư mục đó 60 lần mỗi phút. 

- Bạn có thể tính trung bình nó ra thành 1 thao tác mỗi giây. Với ví dụ đơn giản này, bạn có thể thấy rằng mỗi ứng dụng có yêu cầu IOPS khác nhau.

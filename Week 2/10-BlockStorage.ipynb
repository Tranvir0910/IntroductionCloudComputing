### **Block Storage**

#### **What is Block Storage**

- Lưu trữ Block chia các tệp thành các đơn vị (hoặc block) dữ liệu và lưu trữ mỗi block riêng lẻ dưới một địa chỉ duy nhất. 

- Tương tự như **lưu trữ đính kèm trực tiếp ( Direct Attached Storage )** và **lưu trữ tệp, lưu trữ block** cũng **phải được kết nối với một nút tính toán trước khi có thể sử dụng cho công việc**.

- **Lưu trữ block, giống như lưu trữ tệp, có thể được gắn kết từ các thiết bị lưu trữ từ xa**, làm cho nó rất **chống chịu với sự cố, và giữ cho dữ liệu an toàn hơn nhờ mã hóa trong quá trình truyền và mã hóa ở trạng thái nghỉ, có sẵn trên những thiết bị này**. 



#### **Block Storage Cont**

- Lưu trữ block **được gắn kết như một ổ đĩa cho các nút tính toán bằng cách sử dụng mạng sợi quang riêng biệt**, qua đó **tín hiệu di chuyển với tốc độ của ánh sáng**. 

- **Mạng sợi quang này đắt hơn so với các mạng Ethernet cung cấp lưu trữ tệp, điều này là một trong những lý do khiến lưu trữ block thường có giá cao hơn**. 

- Tuy nhiên, **do dữ liệu di chuyển nhanh hơn và với tính nhất quán về tốc độ, chúng rất phù hợp cho công việc yêu cầu lưu trữ độ trễ thấp để hoạt động hiệu quả**.

- Đối với khối lượng công việc, quan trọng lưu ý rằng, **khác với lưu trữ tệp, có thể gắn kết lên đến 80 nút tính toán trở lên, lưu trữ block thường chỉ được gắn kết cho một nút tính toán duy nhất vào một thời điểm**. 

- **Vì các đĩa chạy ở tốc độ cao và nhất quán**, chúng rất **phù hợp cho các công việc yêu cầu lưu trữ nhanh và ổn định liên tục**, chẳng hạn như **cơ sở dữ liệu và máy chủ thư điện tử**. Lưu trữ block **không phù hợp cho các công việc yêu cầu mức độ chia sẻ đĩa giữa các nút tính toán**.

#### **IOPS**

- Đối với lưu trữ block và tệp, cần **xem xét dung lượng IOPS của lưu trữ**. 

- Hầu hết các nhà cung cấp đám mây **sẽ cho phép bạn chỉ định đặc tính IOPS khi triển khai lưu trữ và trong một số trường hợp**, điều chỉnh IOPS của lưu trữ theo nhu cầu của bạn, để có thể điều chỉnh khi yêu cầu hoặc hành vi sử dụng của ứng dụng thay đổi.

#### **Commonalities**

- Lưu trữ Block và File đến từ các thiết bị được duy trì bởi nhà cung cấp dịch vụ. 

- Cả hai thường có sẵn và chống chịu cao và thường bao gồm mã hóa dữ liệu ở trạng thái nghỉ và trong quá trình truyền.

#### **Differences**

- Lưu trữ Tệp được **kết nối với một nút tính toán bằng cách sử dụng mạng Ethernet**, nên đôi khi được gọi là **Lưu trữ Đính kèm Mạng hoặc Lưu trữ NFS**. 

- Lưu trữ Tệp **rất đáng tin cậy, nhưng tốc độ của mạng kết nối có thể thay đổi dựa trên tải**. Lưu trữ Block **được kết nối thông qua mạng sợi quang tốc độ cao, rất đáng tin cậy và nhất quán**. 

- Lưu trữ Tệp **có thể được kết nối đồng thời với nhiều nút tính toán**. Lưu trữ Block **chỉ có thể được kết nối với một nút vào một thời điểm**. 

- Lưu trữ Tệp **là lựa chọn tốt khi cần các phần chia sẻ tệp, nơi công việc không đòi hỏi kết nối nhanh chóng đến lưu trữ hoặc khi chi phí là một yếu tố quan trọng**. Lưu trữ Block là **lựa chọn tốt khi hỗ trợ một ứng dụng yêu cầu truy cập nhanh và nhất quán đến đĩa, chẳng hạn như cơ sở dữ liệu.** 

- ***Hãy nhớ xem xét yêu cầu IOPS của ứng dụng khi triển khai lưu trữ tệp hoặc block.***

1. **Mục Đích Sử Dụng:**
   - Xác định mục đích sử dụng để quyết định lưu trữ block hay lưu trữ tệp cho ứng dụng của bạn.

2. **VMware và Ổ Đĩa Khởi Động:**
   - Nếu bạn sử dụng VMware với nhiều máy chủ ảo và cần ổ đĩa khởi động, lưu trữ block là lựa chọn phù hợp.

3. **Cơ Sở Dữ Liệu Giao Dịch và Quan Hệ:**
   - Đối với các ứng dụng như cơ sở dữ liệu giao dịch hoặc quan hệ, đòi hỏi độ trễ thấp và hiệu suất cao, lưu trữ block là sự chọn lựa đúng đắn.

4. **Kết Hợp Dữ Liệu Cấu Trúc và Không Cấu Trúc:**
   - Trong trường hợp có cả dữ liệu có cấu trúc và không cấu trúc, như máy chủ lưu trữ web, lưu trữ tệp là sự lựa chọn phù hợp.

5. **Không Gian Cộng Tác và Nhiều Người Dùng:**
   - Nếu cần một không gian cộng tác với nhiều người dùng truy cập, đọc và ghi cùng một lúc, lưu trữ tệp là lựa chọn phù hợp.

6. **Tổng Kết:**
   - Lưu trữ block và lưu trữ tệp là những lựa chọn truyền thống nhưng vẫn rất phổ biến và hữu ích cho các loại công việc khác nhau, cả trên nơi và trong đám mây.

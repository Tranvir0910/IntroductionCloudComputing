### **Object Storage - Tiers and APIs**

#### **Object Storage Tiers**

- Các "bucket" Lưu trữ Đối tượng cũng **có các cấp độ lưu trữ, hoặc các loại, được liên kết với chúng và các cấp độ này dựa trên tần suất truy cập dữ liệu**.

#### **Standard Tier ( Cấp độ tiêu chuẩn )**

**"Bucket" thuộc cấp độ tiêu chuẩn** là nơi bạn sẽ **lưu trữ các đối tượng được truy cập thường xuyên**. Cấp độ này **thường có chi phí cao nhất cho mỗi gigabyte liên quan đến nó**. 

#### **Vault/Archive Tier ( Cấp độ lưu trữ lâu dài )**

**"Vault" hoặc cấp độ lưu trữ lâu dài** là nơi bạn có thể **lưu trữ các tài liệu mà có lẽ chỉ được truy cập một hoặc hai lần mỗi tháng, hoặc ít hơn, và điều này sẽ có chi phí lưu trữ thấp hơn**.

#### **Cold Vault Tier ( Cấp độ lưu trữ lâu dài )**

Cấp độ "cold vault", nơi bạn sẽ **lưu trữ dữ liệu thường chỉ được truy cập một hoặc hai lần mỗi năm**. Chi phí lưu trữ này **thường chỉ là một phần nhỏ của một xu Mỹ mỗi gigabyte mỗi tháng**. 

#### **Automatic Archiving Rules**

Thường, bạn cũng có thể **thiết lập các quy tắc tự động lưu trữ cho dữ liệu** của bạn, có nghĩa là **nếu một đối tượng không được truy cập trong một khoảng thời gian, nó sẽ tự động được chuyển đến một cấp độ lưu trữ rẻ hơn**.

#### **OS - Speed**

- Quy tắc sử dụng một số siêu dữ liệu của đối tượng để xác định khi nào nên được lưu trữ dự trữ. 

- Lưu ý rằng, Lưu trữ Đối tượng **không đi kèm với tùy chọn IOPS**. 

- Lưu trữ Đối tượng thường **rất chậm so với lưu trữ tệp hoặc lưu trữ khối, nơi việc tải xuống thường mất vài giây nếu không phải lâu hơn để hoàn thành**. 

- Khi các nhà cung cấp cung cấp **"bucket" dạng "cold vault"**, việc truy xuất dữ liệu từ những cấp độ này đôi khi **có thể mất thậm chí cả giờ vì lưu trữ được giữ offline**. 

- Nếu ứng dụng của **bạn cần truy cập nhanh chóng đến các tệp, thì Lưu trữ Đối tượng có thể không phải là một lựa chọn tốt**.

#### **OS - Costs**

- Chúng ta đã đề cập rằng giá của Lưu trữ Đối tượng được **tính theo giá mỗi gigabyte sử dụng**, nhưng cũng **có thể có các chi phí khác liên quan đến việc truy xuất dữ liệu**. 

- Những chi phí này tương đối thấp, nhưng **phí truy cập có thể cao hơn đối với dữ liệu ở cấp độ "vault" hoặc "cold vault"**, vì vậy quan trọng là **đảm bảo rằng dữ liệu đang ở cấp độ đúng dựa trên tần suất truy cập của nó**. 

#### **API**

- Lưu trữ Đối tượng không cần được kết nối với một nút tính toán để bạn có thể truy cập nó, thay vào đó, bạn truy cập lưu trữ đối tượng thông qua một giao diện lập trình ứng dụng, hoặc API.

- API **phổ biến nhất cho lưu trữ đối tượng được gọi là API S3**, đó là một **tiêu chuẩn dựa trên lưu trữ đối tượng S3 được cung cấp bởi AWS**. 

- Nhiều **nhà cung cấp cung cấp các API cho lưu trữ đối tượng của họ, tương thích với S3**, điều này rất hữu ích, vì nó có **nghĩa là nhà phát triển có thể viết mã có thể truy cập nhiều giải pháp lưu trữ đối tượng từ nhiều nhà cung cấp khác nhau**. 

- API chính là **một API dựa trên HTTP**, hoặc **dịch vụ web RESTful**. **Cuộc gọi API cho phép ứng dụng quản lý lưu trữ đối tượng và "bucket", cũng như đặt, tải lên hoặc tải xuống các đối tượng từ và vào chúng**. 

#### **OS - Backup solutions**

- Lưu trữ Đối tượng **không chỉ dành cho các ứng dụng mới, mà còn có thể được sử dụng để đáp ứng yêu cầu cho những ứng dụng hiện có**. 

- Nó cũng có thể được **sử dụng như một giải pháp hiệu quả cho sao lưu** và **khôi phục sau thảm họa thay vì sử dụng các giải pháp bảo lưu dựa trên băng cassettes ngoại vi**, **giảm thời gian để khôi phục dữ liệu**. 

- Nhiều gói sao lưu hiện nay đã **bao gồm khả năng sao lưu dữ liệu lên Cloud bằng lưu trữ đối tượng**. 

- Lưu trữ Đối tượng **hiệu quả hơn so với các giải pháp sao lưu băng**, mà **yêu cầu băng cassettes phải được tải vào và lấy ra từ đĩa băng và di chuyển ra khỏi nơi an toàn địa lý**.

#### **Summary**

- **Cấp Độ và Chi Phí:**
  Lưu trữ đối tượng có các cấp độ khác nhau, với chi phí biến động tùy thuộc vào từng cấp độ. Một số cấp độ dựa trên tần suất truy cập đối tượng.

- **Giá Cả và Truy Xuất Dữ Liệu:**
  Lưu trữ đối tượng được tính giá theo mỗi gigabyte sử dụng mỗi tháng, kèm theo một số chi phí truy xuất dữ liệu.

- **So Sánh Chi Phí:**
  Lưu trữ đối tượng có chi phí thấp hơn đáng kể so với lưu trữ tệp hoặc lưu trữ khối.

- **Tốc Độ Truy Cập:**
  Lưu trữ đối tượng có tốc độ truy cập thấp so với lưu trữ tệp và lưu trữ khối.

- **Quy Tắc Tự Động Lưu Trữ:**
  Có thể thiết lập quy tắc tự động để chuyển đối tượng sang các cấp độ lưu trữ rẻ hơn khi chúng ít được truy cập.

- **Truy Cập thông qua API:**
  Lưu trữ đối tượng được truy cập thông qua một API. Nhiều nhà cung cấp có API tương thích với S3, giúp nhà phát triển tạo mã hoạt động trên nhiều giải pháp lưu trữ đối tượng.

- **Giải Pháp Sao Lưu và Phục Hồi:**
  Lưu trữ đối tượng trong đám mây cung cấp giải pháp hiệu quả cho sao lưu và phục hồi sau thảm họa.

### **Object Storage - Tiers and APIs**

#### **Object Storage Tiers**

- Cc "bucket" Lu tr i tng cng **c cc cp  lu tr, hoc cc loi, c lin kt vi chng v cc cp  ny da trn tn sut truy cp d liu**.

#### **Standard Tier ( Cp  tiu chun )**

**"Bucket" thuc cp  tiu chun** l ni bn s **lu tr cc i tng c truy cp thng xuyn**. Cp  ny **thng c chi ph cao nht cho mi gigabyte lin quan n n**. 

#### **Vault/Archive Tier ( Cp  lu tr lu di )**

Cp  "cold vault", ni bn s **lu tr d liu thng ch c truy cp mt hoc hai ln mi nm**. Chi ph lu tr ny **thng ch l mt phn nh ca mt xu M mi gigabyte mi thng**. 

#### **Automatic Archiving Rules**

#### **OS - Speed**

#### **API**

- Lu tr i tng khng cn c kt ni vi mt nt tnh ton  bn c th truy cp n, thay vo , bn truy cp lu tr i tng thng qua mt giao din lp trnh ng dng, hoc API.

#### **Summary**

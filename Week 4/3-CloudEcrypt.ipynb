### **Cloud Encryption**

#### **Encryption**

**Mã hóa là gì:**

- Mã hóa được định nghĩa là làm rối dữ liệu theo một cách khiến nó trở nên không thể đọc được. 



#### **Definition**

- Có hai phần trong một hệ thống mã hóa: thuật toán mã hóa và khóa giải mã. 

    - Thuật toán mã hóa định nghĩa các quy tắc theo đó dữ liệu sẽ được biến đổi để trở nên không thể đọc được. 

    - Khóa giải mã định nghĩa cách dữ liệu đã được mã hóa sẽ được biến đổi trở lại dữ liệu có thể đọc được.

#### **Cloud Encryption Services**

**Dịch Vụ Mã Hóa Đám Mây:**

   - Các nhà cung cấp đám mây cung cấp nhiều dịch vụ mã hóa đám mây, từ việc mã hóa chỉ dữ liệu được xác định là nhạy cảm đến mã hóa end-to-end của tất cả dữ liệu tải lên đám mây.
   
   - Dữ liệu được mã hóa khi nhận được, và khóa mã hóa được chuyển cho khách hàng để giải mã dữ liệu khi cần thiết.

**Quản Lý Khóa:**

   - Khóa cần được quản lý một cách an toàn. Nếu bạn mất khóa, bạn sẽ không thể đọc dữ liệu của mình.
   
   - Dữ liệu cần được bảo vệ ở ba trạng thái: **yên tại, trên đường truyền và khi đang sử dụng**.

#### **Data Protection States**

**Mã Hóa ở Trạng Thái Yên Tại ( rest ):**

   - Mã hóa ở trạng thái yên tại bảo vệ dữ liệu khi nó được lưu trữ vật lý trong cơ sở dữ liệu hoặc lớp lưu trữ.

   - Có nhiều lựa chọn mã hóa dữ liệu yên tại tùy thuộc vào ứng dụng và yêu cầu kinh doanh.

**Mã Hóa Trong Quá Trình Truyền:**

   - Mã hóa trong quá trình truyền bảo vệ dữ liệu khi nó được truyền từ một địa điểm này đến một địa điểm khác.

   - Bao gồm việc mã hóa dữ liệu trước khi truyền, xác thực các đầu cuối, và giải mã và xác minh dữ liệu khi đến nơi.

**Mã Hóa Khi Sử Dụng:**

   - Mã hóa khi sử dụng bảo vệ dữ liệu khi nó đang được sử dụng trong bộ nhớ để thực hiện các phép tính.
   
   - Cho phép thực hiện các phép tính trên văn bản đã được mã hóa mà không cần giải mã dữ liệu.

#### **Client or Server-side**

**Mã Hóa Tại Bên Server và Bên Client:**

   - Mã hóa tại bên server xảy ra sau khi lưu trữ đám mây nhận dữ liệu của bạn, nhưng trước khi dữ liệu được ghi vào đĩa và lưu trữ.

   - Mã hóa tại bên server có thể sử dụng khóa mã hóa do khách hàng quản lý hoặc sử dụng dịch vụ quản lý khóa do nhà cung cấp đám mây cung cấp.

   - Mã hóa tại bên client xảy ra trước khi dữ liệu được gửi đến đám mây. Người dùng có thể sử dụng khóa mã hóa và thuật toán không hiển thị cho nhà cung cấp đám mây, làm cho việc giải mã dữ liệu trở nên khó khăn đối với nhà cung cấp đám mây.

**Chiến Lược Bảo Vệ Dữ Liệu Đa Đám Mây:**

   - Do hầu hết doanh nghiệp hiện nay hoạt động trong môi trường đa đám mây, có nhu cầu triển khai một chiến lược bảo vệ dữ liệu đồng nhất trên các triển khai on-premise, hybrid và đa đám mây.

**Dịch Vụ Mã Hóa Dữ Liệu Đa Đám Mây:**

   - Một số nhà cung cấp đám mây cung cấp dịch vụ mã hóa dữ liệu đa đám mây với một loạt các tính năng như quản lý truy cập dữ liệu, quản lý khóa tích hợp và mã hóa phức tạp kết hợp để cung cấp khả năng mở rộng và linh hoạt để bảo vệ các công việc làm nhạy cảm nhất trong doanh nghiệp, bất kể dữ liệu đang ở đâu.

#### **Key Management**

**Quản Lý Khóa:**

   - Mã hóa không loại bỏ rủi ro an ninh dữ liệu, nhưng nó tách rủi ro an ninh khỏi dữ liệu chính bằng cách chuyển an ninh đến các khóa mã hóa.

   - Khóa này cần được quản lý và bảo vệ chống lại mối đe dọa để giữ cho dữ liệu an toàn.

#### **Key Management Services**

**Dịch Vụ Quản Lý Khóa:**

   - Các dịch vụ quản lý khóa của một số nhà cung cấp đám mây giúp thực hiện quản lý vòng đời cho các khóa mã hóa được sử dụng trong các dịch vụ đám mây hoặc ứng dụng do khách hàng xây dựng.
   
   - Chúng cho phép khách hàng mã hóa dữ liệu nhạy cảm ở trạng thái yên tại và dễ dàng tạo và quản lý toàn bộ vòng đời của các khóa mật mã được sử dụng để mã hóa dữ liệu.

#### **Key Management Best Practices**

**Thực Hành Tốt Nhất Cho Quản Lý Khóa Mã Hóa:**

   - Lưu trữ khóa mã hóa tách biệt từ dữ liệu đã được mã hóa.

   - Sao lưu khóa và kiểm tra thường xuyên.

   - Đổi khóa định kỳ.
   
   - Triển khai xác thực đa yếu tố cho cả khóa chính và khóa phục hồi.
